```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

file_path = "../DS_kaggle_edited_v2/PM2.5(2016).csv"
df = pd.read_csv(file_path)

print(df.head())
```

```python
pm = pd.read_csv(
    "../DS_kaggle_edited_v2/Bangkok/Train/Bangkok.txt",
    comment="%",
    sep="\t",
    header=None,
    names=["year", "month", "day", "hour", "pm25", "pm10_mask", "retrospective"]
)
pm.head()
```

```python
pm["datetime"] = pd.to_datetime(pm[["year","month","day","hour"]])
pm = pm.sort_values("datetime")
pm.head()
```

```python
temp = pd.read_csv(
    "../DS_kaggle_edited_v2/Bangkok/Train/3H_temperature_Bangkok.csv"
)
wind = pd.read_csv(
    "../DS_kaggle_edited_v2/Bangkok/Train/3H_wind_Bangkok.csv"
)

temp.head()
```

```python
temp["datetime"] = pd.to_datetime(temp["datetime"])
wind["datetime"] = pd.to_datetime(wind["datetime"])

pm["datetime"] = pd.to_datetime(pm[["year","month","day","hour"]])
pm["datetime"] = pm["datetime"] + pd.Timedelta(hours=7)

df = pm.merge(temp[["datetime","Temp(C)"]], on="datetime", how="left")
df = df.merge(wind[["datetime","Wind Speed(km/h)"]], on="datetime", how="left")

df.head()
```

```python
df.info()
df.isna().sum()
```

```python
# Use the merged Bangkok data from cells 3-6
# df already contains: year, month, day, hour, pm25, Temp(C), Wind Speed(km/h)
print("Using merged Bangkok data")
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()[0:15]}")
```

```python
# Remove rows with missing values in key columns
df_clean = df[["hour", "Temp(C)", "Wind Speed(km/h)", "pm25"]].dropna()
print(f"Cleaned data shape: {df_clean.shape}")
```

```python
# Display the cleaned data
df_clean.head()
```

```python
print("Ready for modeling with columns:", df_clean.columns.tolist())
```

```python
X = df_clean[["hour", "Temp(C)", "Wind Speed(km/h)"]]
y = df_clean["pm25"]
print(f"Features shape: {X.shape}, Target shape: {y.shape}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

nn_model = MLPRegressor(
    hidden_layer_sizes=(64, 32),
    activation='relu',
    solver='adam',
    max_iter=500,
    random_state=42
)

nn_model.fit(X_train_scaled, y_train)

print("Training complete")
```

```python
from sklearn.metrics import mean_absolute_error

pred = nn_model.predict(X_val_scaled)
mae = mean_absolute_error(y_val, pred)

print("MAE:", mae)
```

```python
import joblib

joblib.dump(nn_model, "nn_model.pkl")
joblib.dump(scaler, "scaler.pkl")

print("Model saved!")
```
